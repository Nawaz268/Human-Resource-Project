```python
#Convert Bools to '1' and '0'

employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x:1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x:1 if x == 'Y' else 0)
```

```python
employee_df
```

```python
#Check for missing values using Seaborn Heatmap
sns.heatmap(employee_df.isnull(), yticklabels = False, cbar = False, cmap = 'Blues')
#Heatmap shows no values which is good, it means the data has no missing values
```

```python
#Creating a histogram
employee_df.hist(bins = 30, figsize = (20,20), color='r')
```

```python
#Dropping 'EmployeeCount', 'Standardhours' , 'Over18' & 'EmployeeNumber' since they do not change and are not valuable information

employee_df.drop(['EmployeeCount','StandardHours', 'Over18', 'EmployeeNumber'], axis = 1, inplace = True)
```

```python
employee_df
```

```python
#Checking how many employees left the company and how many employees stayed back
left_df = employee_df[employee_df['Attrition'] ==1]
stayed_df=employee_df[employee_df['Attrition'] ==0]

print('Total Employees = ', len(employee_df), '\nTotal Employees who left = ', len(left_df), ' In % = ', (len(left_df)/len(employee_df)*100),'\nTotal Employees who stayed = ', len(stayed_df), ' In % = ', (len(stayed_df)/len(employee_df)*100))
```

```python
#Further investigation into those employees who left

left_df.describe()
```

```python
#We can make a few observations, the employees who left tend to be younger
#Employees who stayed have a higher daily rate salary than those who left, showcasing pay is a major factor
#Also, the hourly rate is similar for both employees, indicating more hours for those who stayed, utilization being key
```

```python
#Let us now look at correlation in the dataset
correlations = employee_df.corr()
f, ax = plt.subplots(figsize = (20,20))
ax.tick_params(right=True, top=True, labelright=False, labeltop=True)
sns.heatmap(correlations, annot=True)
```

```python
#Countplot showing the age w.r.t count who left and stayed

plt.figure(figsize = [25,12])
ax = sns.countplot(x = 'Age', hue= 'Attrition', data = employee_df)
legend_labels, _= ax.get_legend_handles_labels()
ax.legend(legend_labels, ['Employees Stayed','Employees Left'], bbox_to_anchor=(1,1), title='Legend')
```

KDE Plots on Continuous Variables

```python
#Analysis on Continuous Variables

#Analysis on Distance from home using a KDE Plot

plt.figure(figsize = (12,7))
sns.kdeplot(left_df['DistanceFromHome'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['DistanceFromHome'], label = 'Employees who stayed', shade = True, color = 'b')
plt.xlabel('Distance From Home')
```

```python
#Analysis on TotalWorkingYears using a KDE Plot

plt.figure(figsize = (12,7))
sns.kdeplot(left_df['TotalWorkingYears'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['TotalWorkingYears'], label = 'Employees who stayed', shade = True, color = 'b')
plt.xlabel('Total Working Years')
```

```python
#Analysis on Years with current manager using a KDE Plot

plt.figure(figsize = (12,7))
sns.kdeplot(left_df['YearsWithCurrManager'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['YearsWithCurrManager'], label = 'Employees who stayed', shade = True, color = 'b')
plt.xlabel('Years with current manager')
```
